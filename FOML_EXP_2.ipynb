```python
#gowthaman A
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(0)
X = np.random.rand(100) * 10  # 100 values between 0 and 10
y = 2.5 * X + 1.5 + np.random.randn(100)  # y = 2.5x + 1.5 + noise

# Least Squares Method to compute slope (m) and intercept (b)
def least_squares(X, y):
    n = len(X)
    mean_x = np.mean(X)
    mean_y = np.mean(y)

    numerator = np.sum((X - mean_x) * (y - mean_y))
    denominator = np.sum((X - mean_x) ** 2)

    m = numerator / denominator
    b = mean_y - m * mean_x

    return m, b

# Predict and plot
m, b = least_squares(X, y)
y_pred = m * X + b

print("Slope (m):", m)
print("Intercept (b):", b)

# Plotting
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, y_pred, color="red", label="Fitted Line")
plt.title("Simple Linear Regression using Least Squares")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
